```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.patches as mpatches
```

```python
file_paths = {
    "caso_1": "data/bench_fase_a_summary.csv",
    "caso_2": "data/bench_fase_b_summary.csv",
    "caso_3": "data/bench_fase_c_summary.csv"
}

cores = ["#094c90", "#90091d"]

sns.set_style("whitegrid")
```

## Caso 1 - Inserção

```python
df1 = pd.read_csv(file_paths["caso_1"])
agg1 = df1.groupby("estrutura").mean(numeric_only=True)
```

```python
agg1_reset = agg1.reset_index()

sns.barplot(data=agg1_reset, x='estrutura', y='avg_tempo_ms', palette=cores)
plt.title("Tempo médio de inserção (Caso 1)")
plt.ylabel("Tempo (ms)")
plt.xticks(rotation=0)
plt.savefig('images/caso1-tempo_medio_insercao.png')
plt.show()

structures = agg1.index
esq = agg1['avg_rot_esq']
dir_rot = agg1['avg_rot_dir']
x = np.arange(len(structures))
width = 0.35

fig, ax = plt.subplots()
bars1 = ax.bar(x - width/2, esq, width, color=['#094c90' if s.lower()=='avl' else '#90091d' for s in structures])
bars2 = ax.bar(x + width/2, dir_rot, width, color=['#4a6fa5' if s.lower()=='avl' else '#b84d5a' for s in structures])
ax.set_xticks(x)
ax.set_xticklabels(structures)
plt.title("Rotações médias na inserção (Caso 1)")
plt.ylabel("Quantidade de rotações")

handles = [
    mpatches.Patch(color='#094c90', label='Rot Esq'),
    mpatches.Patch(color='#4a6fa5', label='Rot Dir'),
    mpatches.Patch(color='#90091d', label='Rot Esq'),
    mpatches.Patch(color='#b84d5a', label='Rot Dir')
]
plt.legend(handles=handles, loc='lower center', bbox_to_anchor=(0.5, -0.15), ncol=4)
plt.savefig('images/caso1-rotacoes_medias_insercao.png')
plt.show()
```
